```python
# Import packages  
import numpy as np
import pandas as pd
```

```python
# Create series from 1 to 100 incremented by 1
one_to_hundred=pd.Series(np.arange(1,101,1))
one_to_hundred
```

```python
# Add the name to th ecolumn, and create a dataframe
columns = ['age']
df = pd.DataFrame(one_to_hundred, columns=columns)
df
```

```python
# Create bins
df['age_grp'] = pd.cut(df['age'], bins = [0,17,24,34,44,54,64,float('Inf')], 
                       labels = [0,
                                 1,
                                 2,
                                 3,
                                 4,
                                 5,
                                 6])
df
```

```python
# Use one hot encoder to define the object
from sklearn import preprocessing
ohe = preprocessing.OneHotEncoder()
```

```python
# Create an array
ohe.fit_transform(df["age_grp"].values.reshape(-1, 1)).toarray()
```

```python
# Create a dataframe from array, note the df variable is embedded
enc_df = pd.DataFrame(ohe.fit_transform(df["age_grp"].values.reshape(-1, 1)).toarray())
enc_df
```

```python
# This create a join
df_encoded = df.join(enc_df)
df_encoded
```

```python
# Let's go ahead and dummy code rename the new dataframe
age_gen_types = ('0-17','18-24','25-34','35-44','45-54','55-64', '65+')
age_gen_types
df_age_gen = pd.DataFrame(age_gen_types, columns=['age_gen_types'])
df_age_gen
df_dummy = pd.get_dummies(df_age_gen, columns=["age_gen_types"], prefix=["Type_is"] )
df_dummy
print(df_dummy.columns)
df_labeled_dummy = df_encoded.join(df_dummy, on = "age_grp")
df_labeled_dummy 
```

```python
# Useful to see how the data frame is dummy coded
df_dummy
```

```python
# We do not need two columns representing the same indicator variable
df_new = df_labeled_dummy.drop(
    [
        0,
        1,
        2,
        3,
        4,
        5,
        6
    ],
    axis=1
)
df_new
```
